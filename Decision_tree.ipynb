```python
ax = plt.axes()

plt.figure(figsize=(10, 10))

pred = best_model_tree.predict(x_val_1)
X = [i for i in range(0,len(y_val_1))]

TEST = 10**y_val_1
PRED = 10**pred

ax.scatter(X, TEST, label = 'True')
ax.scatter(X, PRED, label = 'Prediction')
ax.plot([0,10], [1.91/2]*2, label = 'Threshold')

ax.set_ylabel("WDR Depth of Penetration",{'fontsize':10})
ax.set_xlabel('Samples', {'fontsize':10}) 
#ax.legend()
plt.show()

print(f'Mean Absolute Error: {mean_absolute_error(TEST, pred)} corresponds to {mean_absolute_error(TEST, PRED)/0.15} tiles')

val = []
prediction = []

for i in TEST:
  if i >= 1.91/2:
    val.append(1)
  else:
    val.append(0)

for i in PRED:
  if i >= 1.91/2:
    prediction.append(1)
  else:
    prediction.append(0)

#print(val)
#print(prediction)

TP = 0
TN = 0
FP = 0
FN = 0

acc_counter = 0

for i in range(len(val)):
  if val[i] == 1 and prediction[i] == 1:
    TP += 1
    acc_counter += 1
  elif val[i] == 0 and prediction[i] == 0:
    TN += 0
    acc_counter += 1
  elif val[i] == 1 and prediction[i] == 0:
    FN += 1
  else:
    FP += 1

print(f'Recall: {(TP/(TP+FN))*100}%')
print(f'Accuracy: {(acc_counter/len(val))*100}%')
print(f'Precision: {(TP/(TP+FP))*100}%')
```
